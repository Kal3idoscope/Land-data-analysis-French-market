## Nettoyage des données des Valeurs Foncières 2024-S1 

```python
#### EXPLORING THE DATASET ####
import pandas as pd

# Read in the data
df_2024 = pd.read_csv('ValeursFoncieres-2024-S1.txt', delimiter='|')

df_2024.head()
```

```python
#### BASIC INFO ABOUT THE DATA ####
df_2024.info()
```

```python
#### SHOW IF THERE ARE MISSING VALUES ####
df_2024.isnull().sum()
```

```python
#### DELETE THE COLUMNS WHERE THERE ARE MORE THAN 50% MISSING VALUES ####

threshold = len(df_2024)*0.5
df_2024 = df_2024.dropna(axis=1, thresh = threshold)
df_2024.head()
```

```python
#### CHECK IF VALEUR FONCIERE IS NULL ####
df_2024.isnull().sum()
```

```python
#### DELETE ROWS WHERE 'VALEUR FONCIERE' IS NULL ####
#df_2024.dropna(subset=['Valeur fonciere'], inplace=True)
#df_2024.head()
```

```python
#### DELETE COLUMNS WE DONT WANT TO ANALYSE ####
df_2024.drop(['No voie', 'Type de voie', 'Nature culture',  'Section',  'No plan', 'No disposition', 'Code voie'], axis=1, inplace=True)
df_2024.head()
```

```python
#### REPLACE MISSING VALUES IN COLUMNS WE WANT ####
df_2024.fillna(0, inplace=True)
df_2024.head()
```

```python
#### CHECK TYPE ####
df_2024.info()
```

```python
#### CONVERTS TO THE RIGHT TYPE ####

category_convert = ['Nature mutation', 'Voie', 'Code postal', 'Commune', 'Code departement', 'Code commune', 'Code type local', 'Type local']
df_2024[category_convert] = df_2024[category_convert].astype('category')

df_2024['Date mutation'] = pd.to_datetime(df_2024['Date mutation'], errors = 'coerce')

df_2024['Valeur fonciere']=df_2024['Valeur fonciere'].replace({'\xa0':'',',':'.'}, regex=True)
df_2024['Valeur fonciere']=pd.to_numeric(df_2024['Valeur fonciere'], errors='coerce')

df_2024.info()
```

```python
#### CHECK FOR DUPLICATES ####
df_2024.duplicated().sum()
```

```python
#### DELETE DUPLICATES ####

df_2024.drop_duplicates(keep='last', inplace=True)
df_2024.duplicated().sum()
```
