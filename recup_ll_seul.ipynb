## Nettoyage des données géolocalisées

```python
#### EXPLORING THE DATASET ####
import pandas as pd

# Read in the data
df_ll_s = pd.read_csv('full.csv')

df_ll_s.head()
```

```python
#### BASIC INFO ABOUT THE DATA ####
df_ll_s.info()
```

```python
#### SHOW IF THERE ARE MISSING VALUES ####
df_ll_s.isnull().sum()
```

```python
#### DELETE THE COLUMNS WHERE THERE ARE MORE THAN 50% MISSING VALUES ####

threshold = len(df_ll_s)*0.5
df_ll_s = df_ll_s.dropna(axis=1, thresh = threshold)
df_ll_s.head()
```

```python
#### CHECK IF VALEUR FONCIERE IS NULL ####
df_ll_s.isnull().sum()
```

```python
#### DELETE ROWS WHERE 'VALEUR FONCIERE' IS NULL ####
df_ll_s.dropna(subset=['valeur_fonciere'], inplace=True)
df_ll_s.head()
```

```python
#### DELETE COLUMNS WE DONT WANT TO ANALYSE ####
df_ll_s = df_ll_s[['code_postal', 'latitude', 'longitude']]
df_ll_s.head()
```

```python
#### REPLACE MISSING VALUES IN COLUMNS WE WANT ####
df_ll_s.fillna(0, inplace=True)
df_ll_s.head()
```

```python
#### CHECK TYPE ####
df_ll_s.info()
```

```python
#### CONVERTS TO THE RIGHT TYPE ####
df_ll_s['code_postal'] = df_ll_s['code_postal'].astype(int).astype(str)

df_ll_s.info()
```

```python
#### CHECK FOR DUPLICATES ####
df_ll_s.duplicated().sum()
```

```python
#### DELETE DUPLICATES ####

df_ll_s.drop_duplicates(keep='last', inplace=True)
df_ll_s.duplicated().sum()
```

```python
### RENAME COLUMNNS ####
df_ll_s.rename(columns={'code_postal': 'Code postal', 'latitude': 'Latitude', 'longitude': 'Longitude'}, inplace=True)
```
