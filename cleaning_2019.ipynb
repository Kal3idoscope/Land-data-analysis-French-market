## Nettoyage des données des Valeurs Foncières 2019-S2 

```python
#### EXPLORING THE DATASET ####
import pandas as pd

# Read in the data
df_2019 = pd.read_csv('ValeursFoncieres-2019-S2.txt', delimiter='|')

df_2019.head()
```

```python
#### BASIC INFO ABOUT THE DATA ####
df_2019.info()
```

```python
#### SHOW IF THERE ARE MISSING VALUES ####
df_2019.isnull().sum()
```

```python
#### DELETE THE COLUMNS WHERE THERE ARE MORE THAN 50% MISSING VALUES ####

threshold = len(df_2019)*0.5
df_2019 = df_2019.dropna(axis=1, thresh = threshold)
df_2019.head()
```

```python
#### CHECK IF VALEUR FONCIERE IS NULL ####
df_2019.isnull().sum()
```

```python
#### DELETE ROWS WHERE 'VALEUR FONCIERE' IS NULL ####
df_2019.dropna(subset=['Valeur fonciere'], inplace=True)
df_2019.head()
```

```python
#### DELETE COLUMNS WE DONT WANT TO ANALYSE ####
df_2019.drop(['No voie', 'Type de voie', 'Nature culture',  'Section',  'No plan', 'No disposition', 'Code voie'], axis=1, inplace=True)
df_2019.head()
```

```python
#### REPLACE MISSING VALUES IN COLUMNS WE WANT ####
df_2019.fillna(0, inplace=True)
df_2019.head()
```

```python
#### CHECK TYPE ####
df_2019.info()
```

```python
#### CONVERTS TO THE RIGHT TYPE ####

category_convert = ['Nature mutation', 'Voie', 'Code postal', 'Commune', 'Code departement', 'Code commune', 'Code type local', 'Type local']
df_2019[category_convert] = df_2019[category_convert].astype('category')

df_2019['Date mutation'] = pd.to_datetime(df_2019['Date mutation'], errors = 'coerce')

df_2019['Valeur fonciere']=df_2019['Valeur fonciere'].replace({'\xa0':'',',':'.'}, regex=True)
df_2019['Valeur fonciere']=pd.to_numeric(df_2019['Valeur fonciere'], errors='coerce')

df_2019.info()
```

```python
#### CHECK FOR DUPLICATES ####
df_2019.duplicated().sum()
```

```python
#### DELETE DUPLICATES ####

df_2019.drop_duplicates(keep='last', inplace=True)
df_2019.duplicated().sum()
```
