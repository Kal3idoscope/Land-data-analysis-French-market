```python
#### EXPLORING THE DATASET ####
import pandas as pd

# Read in the data
df_2021 = pd.read_csv('ValeursFoncieres-2021.txt', delimiter='|')

# fnlwgt is a weight variable that represents the demographic characteristics of each person surveyed.
df_2021.head()
```

```python
#### BASIC INFO ABOUT THE DATA ####
df_2021.info()
```

```python
#### SHOW IF THERE ARE MISSING VALUES ####
df_2021.isnull().sum()
```

```python
#### DELETE THE COLUMNS WHERE THERE ARE MORE THAN 50% MISSING VALUES ####

threshold = len(df_2021)*0.5
df_2021 = df_2021.dropna(axis=1, thresh = threshold)
df_2021.head()
```

```python
#### DELETE COLUMNS WE DONT WANT TO ANALYSE ####
df_2021.drop(['No voie', 'Type de voie', 'Nature culture',  'Section',  'No plan', 'No disposition', 'Code voie'], axis=1, inplace=True)
df_2021.head()
```

```python
#### REPLACE MISSING VALUES IN COLUMNS WE WANT ####
df_2021.fillna(0, inplace=True)
df_2021.head()
```

```python
df_2021.isnull().sum()
```

```python
#### CHECK TYPE ####
df_2021.info()
```

```python
#### CONVERTS TO THE RIGHT TYPE ####

category_convert = ['Nature mutation', 'Voie', 'Code postal', 'Commune', 'Code departement', 'Code commune', 'Code type local', 'Type local']
df_2021[category_convert] = df_2021[category_convert].astype('category')

df_2021['Date mutation'] = pd.to_datetime(df_2021['Date mutation'], errors = 'coerce')

df_2021['Valeur fonciere']=df_2021['Valeur fonciere'].replace({'\xa0':'',',':'.'}, regex=True)
df_2021['Valeur fonciere']=pd.to_numeric(df_2021['Valeur fonciere'], errors='coerce')

df_2021.info()
```

```python
#### CHECK FOR DUPLICATES ####
df_2021.duplicated().sum()
```

```python
df_2021.drop_duplicates(keep='last', inplace=True)
df_2021.duplicated().sum()
```
