```python
#### DELETE THE COLUMNS WHERE THERE ARE MORE THAN 50% MISSING VALUES ####

threshold = len(df_2022)*0.5
df_2022 = df_2022.dropna(axis=1, thresh = threshold)
df_2022.head()
```

```python
df_2022.isnull().sum()
```

```python
#### DELETE ROWS WHERE 'VALEUR FONCIERE' IS NULL ####
df_2022.dropna(subset=['Valeur fonciere'], inplace=True)
df_2022.head()
```

```python
#remplacement 
#df_2022 = df_2022.replace('NAN', 'none')

#suppression de colonnes avec +50% de NAN
#colonnes_a_supp = pourcentage_na[pourcentage_na>50].index
#df = df.drop(columns=colonnes_a_supp)

#colonnes_a_supp.shape

#### DELETE COLUMNS WE DONT WANT TO ANALYSE ####
df_2022.drop(['No voie', 'Type de voie', 'Nature culture',  'Section',  'No plan', 'No disposition', 'Code voie'], axis=1, inplace=True)
df_2022.head()
```

```python
df_2022.isnull().sum()
```

```python
#### REPLACE MISSING VALUES IN COLUMNS WE WANT ####
df_2022.fillna(0, inplace=True)
df_2022.head()
```

```python
#### CHECK FOR DUPLICATES ####
df_2022.drop_duplicates(keep='last', inplace=True)
df_2022.duplicated().sum()
```

```python
df_2022.info()
```

```python
#### CONVERTS TO THE RIGHT TYPE ####

category_convert = ['Nature mutation', 'Voie', 'Code postal', 'Commune', 'Code departement', 'Code commune', 'Code type local', 'Type local']
df_2022[category_convert] = df_2022[category_convert].astype('category')

df_2022['Date mutation'] = pd.to_datetime(df_2022['Date mutation'], errors = 'coerce')

df_2022['Valeur fonciere']=df_2022['Valeur fonciere'].replace({'\xa0':'',',':'.'}, regex=True)
df_2022['Valeur fonciere']=pd.to_numeric(df_2022['Valeur fonciere'], errors='coerce')

df_2022.info()
```
